```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as ses
import missingno as ms
```

#### Load data set

```python
data = pd.read_csv('data.csv')
data.info()
```

```python
data.head()
```

```python
answer = pd.get_dummies(data['diagnosis'], drop_first=True)
answer.head()
```

```python
data = pd.concat([data, answer], axis=1)
data.info()
```

#### Column중 'Unnamed:32"는 불필요해 보입니다, 또한 label이 되는 diagnosis는 one-hot encoding으로 처리하였으니 삭제해줍니다

```python
data.drop('diagnosis', axis=1, inplace=True)
data.drop('Unnamed: 32', axis=1, inplace=True)
data.info()
```

```python
canser_data = data[data['M']== 1]
nocanser_data = data[data['M'] == 0]
```

```python
plt.figure(figsize=(20,80))

for index, i in enumerate(data.columns):
    if(i == 'id') or (i=='M'):
        continue
    else:
        plt.subplot(15,2,index)
        plt.title(i)
        plt.boxplot((nocanser_data[i], canser_data[i]))
        plt.xticks([1, 2], ['No cancer', 'Cancer'])
```
